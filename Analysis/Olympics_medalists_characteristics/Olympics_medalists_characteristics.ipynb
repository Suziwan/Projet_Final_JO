## Ã‰volution des caractÃ©ristiques physiques des mÃ©daillÃ©s en athlÃ©tisme

* Courbe d'Ã©volution de la taille moyenne des mÃ©daillÃ©s

```python
# Filtrer les donnÃ©es pour les femmes et les hommes mÃ©daillÃ©s en athlÃ©tisme
femmes_medaillees_athletisme = df[(df['Sex'] == 'F') & (df['Sport'] == 'Athletics') & (df['Medal'].notnull()) & (df ['Year'] >= 1948)]
hommes_medailles_athletisme = df[(df['Sex'] == 'M') & (df['Sport'] == 'Athletics') & (df['Medal'].notnull())& (df ['Year'] >= 1948)]

# Calculer la taille moyenne des femmes mÃ©daillÃ©es en athlÃ©tisme par annÃ©e
taille_moyenne_femmes_medaillees_athletisme = femmes_medaillees_athletisme.groupby('Year')['Height'].mean()

# Calculer la taille moyenne des hommes mÃ©daillÃ©s en athlÃ©tisme par annÃ©e
taille_moyenne_hommes_medailles_athletisme = hommes_medailles_athletisme.groupby('Year')['Height'].mean()

# Calculer la ligne de tendance pour les femmes
coeff_femmes = np.polyfit(taille_moyenne_femmes_medaillees_athletisme.index, taille_moyenne_femmes_medaillees_athletisme, 1)
ligne_tendance_femmes = np.polyval(coeff_femmes, taille_moyenne_femmes_medaillees_athletisme.index)

# Calculer la ligne de tendance pour les hommes
coeff_hommes = np.polyfit(taille_moyenne_hommes_medailles_athletisme.index, taille_moyenne_hommes_medailles_athletisme, 1)
ligne_tendance_hommes = np.polyval(coeff_hommes, taille_moyenne_hommes_medailles_athletisme.index)

# CrÃ©er le graphique
plt.plot(taille_moyenne_femmes_medaillees_athletisme.index, taille_moyenne_femmes_medaillees_athletisme, label='Femmes', color='#F0282D')
plt.plot(taille_moyenne_femmes_medaillees_athletisme.index, ligne_tendance_femmes, '--', label='Tendance femmes', color='#F0282D')
plt.plot(taille_moyenne_hommes_medailles_athletisme.index, taille_moyenne_hommes_medailles_athletisme, label='Hommes', color='#980F30')
plt.plot(taille_moyenne_hommes_medailles_athletisme.index, ligne_tendance_hommes, '--', label='Tendance hommes', color='#980F30')

# Ajouter des titres et des Ã©tiquettes
plt.title("Ã‰volution de la taille moyenne des mÃ©daillÃ©s en athlÃ©tisme", fontproperties=open_sans_font_bold)
plt.xlabel("AnnÃ©e", fontproperties=open_sans_font)
plt.ylabel("Taille moyenne en cm", fontproperties=open_sans_font)
plt.legend(prop=open_sans_font)

#Ne pas afficher le cadre du graphe
plt.box(False)

# Afficher le graphique
plt.show()
```
